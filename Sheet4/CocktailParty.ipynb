```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
```

```python
def plot_signals(df, title, stepsize):
    time = list(range(0,df.shape[0], stepsize))
    fig, ax = plt.subplots(2,1)
    fig.suptitle(f'{title}: signals vs time')
    ax[0].plot(time,df.iloc[time,0], markersize=3)
    ax[0].set_ylabel('First Comp')
    ax[1].plot(time,df.iloc[time,1], markersize=3)
    ax[1].set_ylabel('Second Comp')
    ax[1].set_xlabel('Time')

    plt.figure()
    plt.scatter(df.iloc[time,0],df.iloc[time,1], linewidths=0.5)
    plt.title(f'{title}: joint distribution')
    plt.xlabel('First Comp')
    plt.ylabel('Second Comp')
```

```python
toy = pd.read_csv('toydata.txt', delimiter=' ', header=None)
audio = pd.read_csv('audiodata.txt', delimiter=' ', header=None)
```

```python
plot_signals(toy, 'toydata', 1)
plot_signals(audio, 'audiodata', 15)
```

```python
pcaToy = PCA(n_components=2, whiten=True)
toyPCATransformed = pd.DataFrame(pcaToy.fit_transform(toy))
print(pcaToy.components_)
print(pcaToy.explained_variance_ratio_)
```
